```python
import numpy as np
import matplotlib.pyplot as plt

# Dane 
czas_zajetosci = np.array([10, 30, 50, 70, 90, 110, 130, 150, 170])
max_liczba_uzytkownikow = np.array([11437367, 3812455, 2287473, 1633909, 1270818, 1039760, 879797, 762491, 672786])

# Dopasowanie wielomanu
stopien_wielomianu = 3 #DO DOSTOSOWANIA

wspolczynniki = np.polyfit(czas_zajetosci, max_liczba_uzytkownikow, stopien_wielomianu)
aproksymacja_wielomianowa = np.poly1d(wspolczynniki)

# Przygotowanie punktów do wykresu
czas_zajetosci_przewidywane = np.linspace(min(czas_zajetosci), max(czas_zajetosci), 1400)
max_liczba_uzytkownikow_przewidywane = aproksymacja_wielomianowa(czas_zajetosci_przewidywane)

# Wykres dopasowania modelu wielomianowego
plt.figure(figsize=(10, 6))
plt.scatter(czas_zajetosci, max_liczba_uzytkownikow, color='blue', label='Dane rzeczywiste')
plt.plot(czas_zajetosci_przewidywane, max_liczba_uzytkownikow_przewidywane, color='red', linewidth=2, label='Aproksymacja wielomianowa')
plt.xlabel('Czas zajętości')
plt.ylabel('Maksymalna liczba użytkowników')
plt.title('Aproksymacja wielomianowa: liczba abonentów w funkcji zajętości kanału')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Dane B - regresja liniowa
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Dane
pasmo_kHz = np.array([1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500, 7000, 7500, 8000, 8500, 9000, 9500, 10000])
liczba_kanalow = np.array([8, 8, 16, 24, 24, 32, 40, 40, 48, 48, 56, 64, 64, 72, 80, 80, 88, 88, 96])

# Przygotowanie danych do regresji
X = pasmo_kHz.reshape(-1, 1)  # Pasmo jako cecha
y = liczba_kanalow  # Liczba kanałów jako zmienna celu

# Inicjalizacja modelu regresji liniowej
model = LinearRegression()

# Trenowanie modelu na danych
model.fit(X, y)

# Przewidywanie na podstawie modelu
y_pred = model.predict(X)

# Wykres dopasowania modelu
plt.figure(figsize=(10, 6))
plt.scatter(pasmo_kHz, liczba_kanalow, color='blue', label='Dane rzeczywiste')
plt.plot(pasmo_kHz, y_pred, color='red', linewidth=2, label='Regresja liniowa')
plt.xlabel('Pasmo [kHz]')
plt.ylabel('Liczba kanałów')
plt.title('Regresja liniowa: liczba kanałów w systemie a pasmo częstotliwości')
plt.legend()
plt.grid(True)
plt.show()
```
